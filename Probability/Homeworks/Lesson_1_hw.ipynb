**Задание 7**

**В трех углах равностороннего треугольника находится по роботу. Каждый из роботов начинает двигаться в случайно выбранный угол по стороне треугольника. Какова вероятность того, что ни один из роботов не столкнется с другим роботом?**

Ответ: каждый робот может пойти в две стороны, всего роботов три. Значит, число возможных комбинаций направлений таково: 2 × 2 × 2 = 8. Из всех комбинаций только две удовлетворяют условию, что они не встретятся.

Вспоминаем формулу вычисления вероятностей: p = m ÷ n, где m — число исходов, благоприятствующих осуществлению события, а n — число всех равновозможных исходов. Подставим наши цифры: 2 ÷ 8 = ¼. Значит, шанс избежать столкновения — один из четырёх.
