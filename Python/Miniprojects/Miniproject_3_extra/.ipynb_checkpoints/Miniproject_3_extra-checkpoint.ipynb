# Задачи:
* извлечь из таблицы записи с ИНН, указанными в текстовом файле
* записать результат в файл selected_inn.csv

**1. Извлечь из таблицы записи с ИНН, указанными в текстовом файле.**

**NB! У этого набора данных нет заголовка, поэтому вам пригодится аргумент header=None.**

```python
import pandas as pd
```

```python
necessary = []
with open('C:/Users/Александр/pj/karpov_courses/Python/Miniprojects/Miniproject_3_extra/4_necessary_inn.txt') as f:
    for line in f:
        necessary.append(int(line))
```

```python
data = pd.read_excel('C:/Users/Александр/pj/karpov_courses/Python/Miniprojects/Miniproject_3_extra/4_inn.xls', index_col='#')
data.head(2)
```

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
data.head(2)
```

```python
data.shape
```

```python
data = data.loc[data.head_inn.isin(necessary)]
data.shape
```

**2. Записать результат в файл selected_inn.csv**

```python
data.to_csv('selected_inn.csv', index=False)
```
